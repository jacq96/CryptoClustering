```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
columns = df_market_data.columns
market_data_scaled = StandardScaler().fit_transform(df_market_data[columns])
market_data_scaled[:5]
```

```python
index_name = df_market_data.index.name
print(index_name)
```

```python
scaler = StandardScaler()
```

```python
# Create a DataFrame with the scaled data
# Copy the crypto names from the original data
# Set the coinid column as index
market_data_scaled = scaler.fit_transform(df_market_data)
columns = df_market_data.columns
market_data_scaled_df = pd.DataFrame(market_data_scaled, columns=columns, index=df_market_data.index)

# Display sample data
market_data_scaled_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(market_data_scaled_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
elbow_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_df.hvplot.line(
    x="k",
    y="inertia",
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer: I believe the optimal value for k is 4, that is when the decrease in inertia starts slowing down and doesn't have as drastic of rate of change

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(market_data_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
preds = model.predict(market_data_scaled_df)

# Print the resulting array of cluster values.
print(preds)
```

```python
# Create a copy of the DataFrame
market_data_preds_df = market_data_scaled_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
market_data_preds_df['prediction_clusters'] = preds

# Display sample data
market_data_preds_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
market_data_preds_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    hover_cols=['index'],
    color=preds
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
principal_components = pca.fit_transform(df_market_data)

# View the first five rows of the DataFrame. 
principal_components[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
#finding the sum of the explained variance
total_explained_variance = sum(pca.explained_variance_ratio_)
print(total_explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer: 0.9998139161298989

```python
# Create a new DataFrame with the PCA data.
# Creating a DataFrame with the PCA data
# Copy the crypto names from the original data
# Set the coinid column as index
principal_comp_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'], index=df_market_data.index)


# Display sample data
principal_comp_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(principal_comp_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k': k, "inertia": inertia}
# Create a DataFrame with the data to plot the Elbow curve
elbow_pca_df = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_pca_df.hvplot.line(
    x='k',
    y='inertia',
    xticks=k
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer: 2


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer: yes, the rate of change slows at 2 instead of 4 when using the PCA data. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
k_model = KMeans(n_clusters=3)
```

```python
# Fit the K-Means model using the PCA data
k_model.fit(principal_comp_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
clusters = k_model.predict(principal_comp_df)
# Print the resulting array of cluster values.
print(clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
df_with_clusters = principal_comp_df.copy()

# Add a new column to the DataFrame with the predicted clusters
df_with_clusters['cluster'] = clusters

# Display sample data
df_with_clusters.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
df_with_clusters.hvplot.scatter(
    x='PC1',
    y='PC2',
    hover_cols=['coin_id']
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves (original)
scaler = StandardScaler()
k_values = range(1, 11)
original_scaled = scaler.fit_transform(df_market_data)
orig_inertia = []
for k in k_values:
    k_model = KMeans(n_clusters=k, random_state=1)
    k_model.fit(df_market_data)
    orig_inertia.append(k_model.inertia_)
    
orig_elbow_data = {'k': k_values, "inertia": orig_inertia}
original_elbow_df = pd.DataFrame(orig_elbow_data)


#plot data
original_elbow_df.hvplot.line(
    x='k',
    y='inertia',
    xticks=k,
    title='Original Data Elbow Curve'
)
```

```python
# Composite plot to contrast the Elbow curves (PCA)
k_values_pca = range(1, 11)
PCA_scaled = scaler.fit(principal_comp_df)
pca_inertia = []
for k in k_values_pca:
    k_pca = KMeans(n_clusters=k, random_state=1)
    k_pca.fit(principal_comp_df)
    pca_inertia.append(k_pca.inertia_)
    
pca_elbow_data = {'k': k_values_pca, "inertia": pca_inertia}
pca_elbow_df = pd.DataFrame(pca_elbow_data)
    

#plot data
pca_elbow_df.hvplot.line(
    x='k',
    y='inertia',
    xticks=k,
    title='PCA Data Elbow Curve'
)
```

```python
# Composite plot to contrast the clusters (original)
import matplotlib.pyplot as plt

x_original = df_market_data.values
x_pca = principal_comp_df.values

k_original = KMeans(n_clusters=4, random_state=0)
clusters_original = k_original.fit_predict(x_original)

k_pca = KMeans(n_clusters=4, random_state=0)
clusters_pca = k_pca.fit_predict(x_pca)

plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.scatter(x_original[:, 0], x_original[:, 1], c=clusters_original, cmap='viridis', s=50, alpha=0.5)
plt.title('Clusters from Original Data')

plt.subplot(1, 2, 2)
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=clusters_pca, cmap='viridis', s=50, alpha=0.5)
plt.title('Clusters from PCA Data (Transformed)')
```
